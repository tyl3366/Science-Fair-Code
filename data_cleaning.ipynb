```python
import pandas as pd
```

```python
data_file_path = "raw_data.csv"
data = pd.read_csv(data_file_path)
data = data.drop("Unnamed: 13", axis=1)
```

```python
data.head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
num_imputer = SimpleImputer()
```

```python
num = (data.dtypes == 'float64')
number_cols = list(num[num].index)
number_data = data[number_cols]
```

```python
imputed_number_data = pd.DataFrame(num_imputer.fit_transform(number_data))
```

```python
imputed_number_data.columns = number_data.columns
```

```python
imputed_number_data.head()
```

```python
col_imputer = SimpleImputer(strategy='most_frequent')
```

```python
obj = (data.dtypes == 'object')
object_cols = list(obj[obj].index)
object_data = data[object_cols]
```

```python
object_data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
imputed_col_data = pd.DataFrame(col_imputer.fit_transform(object_data))
```

```python
imputed_col_data.head()
```

```python
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols = pd.DataFrame(OH_encoder.fit_transform(imputed_col_data))
```

```python
OH_cols.head()
```

```python
OH_cols.index = object_data.index
```

```python
OH_cols.head()
```

```python
OH_cols.columns = OH_cols.columns.astype(str)
```

```python
imputed_number_data
```

```python
clean_data = pd.concat((imputed_number_data, OH_cols), axis=1)
```

```python
clean_data
```

```python
clean_data.to_csv('clean_data.csv')
```
