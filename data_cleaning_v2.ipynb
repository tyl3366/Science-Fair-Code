```python
import pandas as pd
```

```python
data_file_path = "raw_data_v2.csv"
data = pd.read_csv(data_file_path)
data = data.drop("Unnamed: 0", axis=1)
```

```python
data
```

```python
from sklearn.impute import SimpleImputer
```

```python
num_imputer = SimpleImputer()
```

```python
num = (data.dtypes == 'float64')
number_cols = list(num[num].index)
number_data = data[number_cols]
```

```python
imputed_number_data = pd.DataFrame(num_imputer.fit_transform(number_data))
```

```python
imputed_number_data.columns = number_data.columns
```

```python
imputed_number_data
```

```python
personality_features = ['Extraverted', 
            'Intuitive', 
            'Thinking', 
            'Judging', 
            'Assertive',
            'Age']
```

```python
personality_data = data[personality_features]
```

```python
clean_data = pd.concat((personality_data, imputed_number_data), axis=1)
```

```python
clean_data
```

```python
clean_data.to_csv('clean_data_v2.csv')
```
